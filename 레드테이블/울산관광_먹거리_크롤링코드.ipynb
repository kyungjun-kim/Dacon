```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
import csv 
import re
from bs4 import BeautifulSoup
import unicodedata
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# selenium 이용해서 드라이버 지정하고 url 불러오기
# keyword 를 포함한 url로 쓰기
driver = webdriver.Chrome(r"C:\Users\USER\Downloads\chromedriver.exe")

# 사이트 접속하기
site_url = 'https://tour.ulsan.go.kr/kor/unit/food/list.ulsan?mId=001002009000000000' 
driver.get(site_url)
time.sleep(3)
```

```python
# 상호명이 저장될 리스트 생성
a = []
```

```python
start = time.time()  # 시작 시간 저장
for i in range(1,24) :
    if i!= 1 :
        driver.get('https://tour.ulsan.go.kr/kor/unit/food/list.ulsan?mId=001002009000000000&page={}&picPrsId=&searchType4=&searchType1=&searchTxt='.format(i))
    for j in range(1,10) :
        driver.find_elements(by = By.CSS_SELECTOR, value ='#listForm > div > fieldset > div.actpic_listbox > div > ul > li:nth-child({}) > a > div.actpic_img > img'.format(j))[0].click()
        time.sleep(0.2)
        title = driver.find_elements(by = By.CSS_SELECTOR, value = '#viewForm > div.glist_view > div.glist_inner.view_txtbox > div > div.view_txthead > p.txthead_tit')[0].text
        content = driver.find_elements(by = By.CSS_SELECTOR, value = '#viewForm > div.glist_view > div.glist_inner.view_txtbox > div > div.view_txtbody > ul')[0].text 
        a.append((title.replace('구분','') + content).split('\n'))
        time.sleep(0.2)
        driver.back()
        time.sleep(0.5)
        
print("time :", time.time() - start)  # 현재시각 - 시작시간 = 실행 시간
```
