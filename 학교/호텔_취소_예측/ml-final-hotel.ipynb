```python
sub_sam = pd.read_csv('/kaggle/input/2020-ml-competition/submission_sample.csv')

train = pd.read_csv('/kaggle/input/2020-ml-competition/book_train.csv')
tr = train

test = pd.read_csv('/kaggle/input/2020-ml-competition/book_test.csv')
te = test
```

```python
tr.hotel_type.value_counts()
```

```python
tr.hotel_type = (tr.hotel_type == "C").astype(int)
te.hotel_type = (te.hotel_type == "C").astype(int)
```

```python
#tr 룸타입
room_type_list = list(set(tr.assigned_room_type.values))

j = 0
for i in room_type_list :
    tr.loc[tr["assigned_room_type"] == i, "assigned_room_type"] = j
    j += 1
    
tr.assigned_room_type = tr['assigned_room_type'].astype(int)


# te
room_type_list = list(set(te.assigned_room_type.values))

j = 0
for i in room_type_list :
    te.loc[te["assigned_room_type"] == i, "assigned_room_type"] = j
    j += 1
    
te.assigned_room_type = te['assigned_room_type'].astype(int)
```

```python
tr.book_method.value_counts()
```

```python
book_method_list = list(set(tr.book_method.values))


j = 0
for i in book_method_list :
    tr.loc[tr["book_method"] == i, "book_method"] = j
    te.loc[te["book_method"] == i, "book_method"] = j
    j += 1

tr.book_method = tr['book_method'].astype(int)
te.book_method = te['book_method'].astype(int)


reserved_room_list = list(set(tr.reserved_room_type.values))

j = 0
for i in reserved_room_list :
    tr.loc[tr["reserved_room_type"] == i, "reserved_room_type"] = j
    te.loc[te["reserved_room_type"] == i, "reserved_room_type"] = j
    j += 1

tr.reserved_room_type = tr['reserved_room_type'].astype(int)
te.reserved_room_type = te['reserved_room_type'].astype(int)


customer_type_list = list(set(tr.customer_type.values))

j = 0
for i in customer_type_list :
    tr.loc[tr["customer_type"] == i, "customer_type"] = j
    te.loc[te["customer_type"] == i, "customer_type"] = j
    j += 1

tr.customer_type = tr['customer_type'].astype(int)
te.customer_type = te['customer_type'].astype(int)


month_list = ["January", "February", "March", "April", "May", "June",
              "July","August","September", "October","November","December"]

for i in range(12) :
    tr.loc[tr["month"] == month_list[i], "month"] = int(i+1)
    te.loc[te["month"] == month_list[i], "month"] = int(i+1)
    
tr.month = tr['month'].astype(int)
te.month = te['month'].astype(int)
```

```python
deposit_list = list(set(tr.deposit.values))

j = 0
for i in deposit_list :
    tr.loc[tr["deposit"] == i, "deposit"] = j
    te.loc[te["deposit"] == i, "deposit"] = j
    j += 1

tr.deposit = tr['deposit'].astype(int)
te.deposit = te['deposit'].astype(int)
```

```python
meal_list = ["BB", "HB", "No_meal", "FB"]

for i in range(len(meal_list)) :
    tr.loc[tr["meal"] == meal_list[i], "meal"] = int(i+1)
    te.loc[te["meal"] == meal_list[i], "meal"] = int(i+1)
    
tr.meal = tr['meal'].astype(int)
te.meal = te['meal'].astype(int)
```

```python
tr2 = tr.drop(columns=['seq',"is_canceled", "country"])
```

```python
y = tr.is_canceled
x = tr2
np.random.seed(123)
```

```python
# 로지스틱회귀
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression

clf_log = LogisticRegression(max_iter=200)
clf_log.fit(x_train, y_train)
```

```python
pd.DataFrame({'변수명':x.columns, '계수값':clf_log.coef_[0]})
```

```python
pred_log = clf_log.predict(x_test)
pred_log
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred_log)
```

```python
from sklearn.metrics import roc_auc_score
pred_log2 = clf_log.predict_proba(x_test)[:,1]
roc_auc_score(y_test, pred_log2)
```

```python
from sklearn.metrics import roc_curve

fprs, tprs, threshold = roc_curve(y_test, pred_log2)

plt.plot(fprs, tprs, label='ROC')
plt.plot([0,1],[0,1], '--', label='Random')
plt.legend();
```

```python
from sklearn.model_selection import cross_val_score

clf_log = LogisticRegression(max_iter=200)
scores = cross_val_score(clf_log, x, y, scoring='roc_auc', cv=5)
print(scores)
print('Mean: ', np.mean(scores))
```
